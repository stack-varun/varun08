```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mckinsey.csv')
```

```python
df
```

```python
#add a row
#loc/iloc
#concat()
```

```python
new_row = {'country': 'India', 'year': 2000,'population':13500000, 'continent': "Asia", 'life_exp':37.08, 'gdp_cap':900.23}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
df
```

```python
df.index
```

```python
len(df.index)
```

```python
new_row = {'country': 'India', 'year': 2000,'population':13600000, 'continent': "Asia", 'life_exp':37.08, 'gdp_cap':900.23}
new_row_val = list(new_row.values())
new_row_val
```

```python
df.loc[len(df.index)] = new_row_val
df
```

```python
df.iloc[len(df.index)-1] = ['Japan', 1000, 1350000, 'Asia', 37.08, 100.23]  #iloc modifes, it cannot add a new record
df
```

```python
df = df.drop(4,axis=0)  #for 2D data O index is vertical
df
```

```python
df.loc[5]
```

```python
df.iloc[4]
```

```python
df.iloc[5]   #iloc takes implicit, we cannot drop implicit data, it will update the indexes automatically
```
